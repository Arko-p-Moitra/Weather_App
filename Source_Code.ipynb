```python
from PyQt5 import *
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import QApplication, QMainWindow
from PyQt5.QtGui import *
import requests
import datetime
import sys

class myWindow(QMainWindow):
    
    
    
    
    
    def __init__(self):
        super(myWindow,self).__init__()
        self.setGeometry(400,55 ,500, 650)
        self.setWindowTitle("Weather App")
        self.initUI()
        self.setStyleSheet('background-color:white')
    
    
            
    def initUI(self):
        
        #Enter your city label
        self.firstL = QtWidgets.QLabel(self)
        self.firstL.setText('Enter the city')
        self.firstL.move(30, 15)
        
       
       
        #TextBox to search City
        self.city = QtWidgets.QLineEdit(self)
        self.city.setGeometry(130, 20, 250, 25)
        self.city.setText("")
        
        
        
        #Search Button
        self.searchBtn = QtWidgets.QPushButton(self)
        self.searchBtn.setText('Search')  
        self.searchBtn.setGeometry(390, 19, 75, 23)
        
        #lable
        self.label = QtWidgets.QLabel(self)
        self.label.setText("Today's Weather")
        self.label.setFont(QFont('Avenir', 10)) 
        self.label.setFont(QtGui.QFont("Avenir",weight=QtGui.QFont.Bold))
        self.label.move(190,50)
        
        
        #First day Night Icon
        self.dayNight=QtWidgets.QLabel(self)
        self.dayNight.setGeometry(205,80,70,66)
        self.dayNight.setText("Hiii")
        self.dayNight.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/download.png"))   
        self.dayNight.setScaledContents(True)
        self.dayNight.setObjectName("photo")

        #Temparature & Location
                
        self.tempGui=QtWidgets.QLabel(self)
        self.tempGui.setGeometry(200,150,160,55)
        self.searchBtn.clicked.connect(self.get_city_temp)
        self.tempGui.setFont(QFont('Avenir', 25)) 
        
        self.locationGUI=QtWidgets.QLabel(self)
        self.locationGUI.setGeometry(180,200,160,41)
        self.searchBtn.clicked.connect(self.get_city)
        self.locationGUI.setFont(QFont('Avenir', 16)) 
        
        
         #Today,Tomorrow,other days label
        
        self.day = QtWidgets.QComboBox(self)
        self.day.setGeometry(QtCore.QRect(150, 250, 85, 30))
        self.day.setObjectName("day")
        self.day.addItem("Today")
        self.day.addItem("Tomorrow")
        self.day.addItem("Other")
        
        self.datesearchBtn = QtWidgets.QPushButton(self)
        self.datesearchBtn.setText('Search')  
        self.datesearchBtn.setGeometry(280, 250, 85, 30)
        
        #00:00 Temp,06:00 Temp,12:00 Temp,18:00
        self.midnightTempGUI=QtWidgets.QLabel(self)
        self.midnightTempGUI.setGeometry(50,300,121,160)
        self.midnightTempGUI.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/first.png"))   
        self.midnightTempGUI.setScaledContents(True)
        self.midnightTempGUI.setObjectName("photo")
        
        self.midnighttemptextTime=QtWidgets.QLabel(self.midnightTempGUI)
        self.midnighttemptextTime.setGeometry(40,20,50,20)
        
        self.midnighttemptextTime.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.midnighttemptextTime.setText("<font color='White'>00:00</font>")
        self.midnighttemptextTime.setFont(QtGui.QFont("Avenir",13))
        
        self.midNightTempFortoday=QtWidgets.QLabel(self.midnightTempGUI)
        self.midNightTempFortoday.setGeometry(40,75,50,20)
        
        self.midNightTempFortoday.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.datesearchBtn.clicked.connect(self.get_specific_temp)
        self.midNightTempFortoday.setFont(QtGui.QFont("Avenir",13))
        
        
        
       
        self.eightAMTempGUI=QtWidgets.QLabel(self)
        self.eightAMTempGUI.setGeometry(190,300,121,160)
        self.eightAMTempGUI.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/secondBlock.png"))   
        self.eightAMTempGUI.setScaledContents(True)
        self.eightAMTempGUI.setObjectName("photo")
        
        self.eightAMtemptextTime=QtWidgets.QLabel(self.eightAMTempGUI)
        self.eightAMtemptextTime.setGeometry(40,20,50,20)
        
        self.eightAMtemptextTime.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.eightAMtemptextTime.setText("<font color='White'>08:00</font>")
        self.eightAMtemptextTime.setFont(QtGui.QFont("Avenir",13))
        
        self.eightAMTempFortoday=QtWidgets.QLabel(self.eightAMTempGUI)
        self.eightAMTempFortoday.setGeometry(40,75,50,20)
        
        self.eightAMTempFortoday.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.datesearchBtn.clicked.connect(self.get_specific_temp)
        self.eightAMTempFortoday.setFont(QtGui.QFont("Avenir",13))
        
        
        self.fourPMTempGUI=QtWidgets.QLabel(self)
        self.fourPMTempGUI.setGeometry(337,300,121,160)
        self.fourPMTempGUI.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/thirdBlock.png"))   
        self.fourPMTempGUI.setScaledContents(True)
        self.fourPMTempGUI.setObjectName("photo")
        
        self.fourPMtemptextTime=QtWidgets.QLabel(self.fourPMTempGUI)
        self.fourPMtemptextTime.setGeometry(40,20,50,20)
        
        self.fourPMtemptextTime.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.fourPMtemptextTime.setText("<font color='White'>16:00</font>")
        self.fourPMtemptextTime.setFont(QtGui.QFont("Avenir",13))
        
        self.fourPMTempFortoday=QtWidgets.QLabel(self.fourPMTempGUI)
        self.fourPMTempFortoday.setGeometry(40,75,50,20)
        
        self.fourPMTempFortoday.setStyleSheet('background-color: rgba(0,0,0,0%)')
        self.datesearchBtn.clicked.connect(self.get_specific_temp)
        self.fourPMTempFortoday.setFont(QtGui.QFont("Avenir",13))
       
        
        #Additional Info
        self.additinalInfo=QtWidgets.QLabel(self)
        self.additinalInfo.setText("Additional Info :")
        self.additinalInfo.setFont(QFont('Avenir', 9, QFont.Bold)) 
        self.additinalInfo.move(30,480)
        
        #Other Weather info
        self.lastFrame=QtWidgets.QGroupBox(self)
        self.lastFrame.setGeometry(30,510,450,110)
        self.lastFrame.setObjectName("groupBox")
        self.lastFrame.setStyleSheet('background-color:#f6f6f6')
        
        self.windSpeedIcon=QtWidgets.QLabel(self.lastFrame)
        self.windSpeedIcon.setGeometry(10,37,21,16)
        self.windSpeedIcon.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/143-512.png"))   
        self.windSpeedIcon.setScaledContents(True)
        self.windSpeedIcon.setObjectName("windphoto")
        
        self.windSpeedGUI = QtWidgets.QLabel(self.lastFrame)
        self.windSpeedGUI.setGeometry(QtCore.QRect(36,30,350,30))
        self.windSpeedGUI.setObjectName("windSpeed")
        self.windSpeedGUI.setText('WindSpeed = ')
        self.searchBtn.clicked.connect(self.get_windSpeed)
        self.windSpeedGUI.setFont(QFont('Avenir', 9, QFont.Bold)) 
        
        
        
        self.humidityIconGui=QtWidgets.QLabel(self.lastFrame)
        self.humidityIconGui.setGeometry(4,77,25,19)
        self.humidityIconGui.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/Humidity-icon.png"))   
        self.humidityIconGui.setScaledContents(True)
        self.humidityIconGui.setObjectName("humidityphoto")
        
        self.humidityGUI = QtWidgets.QLabel(self.lastFrame)
        self.humidityGUI.setGeometry(QtCore.QRect(36,70,350,30))
        self.humidityGUI.setObjectName("humidity")
        self.humidityGUI.setText('Humidity =')
        self.searchBtn.clicked.connect(self.get_humid)
        self.humidityGUI.setFont(QFont('Avenir', 9, QFont.Bold)) 
        
        
        
        
        self.feelIconGui=QtWidgets.QLabel(self.lastFrame)
        self.feelIconGui.setGeometry(250,70,25,19)
        self.feelIconGui.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/temperature-thermometer-pngrepo-com.png"))   
        self.feelIconGui.setScaledContents(True)
        self.feelIconGui.setObjectName("feelphoto")
        
        self.tempFeelsLike = QtWidgets.QLabel(self.lastFrame)
        self.tempFeelsLike.setGeometry(QtCore.QRect(280,65,165,35))
        self.tempFeelsLike.setObjectName("tempfeels")
        self.tempFeelsLike.setText('Temp. feels like = ')
        self.searchBtn.clicked.connect(self.feels)
        self.tempFeelsLike.setFont(QFont('Avenir', 9, QFont.Bold)) 
        
        self.pressureIcon=QtWidgets.QLabel(self.lastFrame)
        self.pressureIcon.setGeometry(250,37,25,19)
        self.pressureIcon.setPixmap(QtGui.QPixmap("D:/GUI/API/Images/air-pressure-1424009-1204225.png"))   
        self.pressureIcon.setScaledContents(True)
        self.pressureIcon.setObjectName("feelphoto")
        
        self.pressure= QtWidgets.QLabel(self.lastFrame)
        self.pressure.setGeometry(QtCore.QRect(280,30,140,35))
        self.pressure.setObjectName("pressure")
        self.pressure.setText('Pressure =')
        self.searchBtn.clicked.connect(self.get_pressure)
        self.pressure.setFont(QFont('Avenir', 9, QFont.Bold)) 
      
    def get_city_temp(self):
            
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
            degree_sign = u"\N{DEGREE SIGN}"
            self.tempGui.setText(Wapi.tempNow()+degree_sign)
            
    def get_city(self):
            
            self.locationGUI.setText(self.city.text())
            
    def get_windSpeed(self):
            
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
            self.windSpeedGUI.setText('Windspeed = '+Wapi.windSpeed()+'m/s')
            
    def get_humid(self):
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
            self.humidityGUI.setText('Humidity = '+Wapi.humid()+'%')
    
    def weather_type(self):      
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
            self.currentWType.setText('Weather Type = '+Wapi.weatherType()+'%')
            
    def feels(self):
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
            degree_sign = u"\N{DEGREE SIGN}"
            self.tempFeelsLike.setText('Temp. feels like = '+Wapi.feel()+degree_sign)
            
    def get_pressure(self):       
            
            selctedCity=self.city.text()
            Wapi=SingledayAPI(selctedCity)
           
            self.pressure.setText('Pressure = '+Wapi.get_pressure()+' mbar')
            
    def get_specific_temp(self):        
            
                selctedCity=self.city.text()
                Wapi=SingledayAPI(selctedCity)
                degree_sign = u"\N{DEGREE SIGN}"
                
                x=self.day.currentText()
                
                if x=='Today':
                     self.midnighttemptextTime.setGeometry(40,20,50,20)  
                     self.midnighttemptextTime.setText("<font color='White'>00:00</font>")
                     self.eightAMtemptextTime.setGeometry(40,20,50,20)   
                     self.eightAMtemptextTime.setText("<font color='White'>08:00</font>")
                     self.fourPMtemptextTime.setGeometry(40,20,50,20)  
                     self.fourPMtemptextTime.setText("<font color='White'>16:00</font>")
                     self.midNightTempFortoday.setText("<font color='White'>"+Wapi.todaymidnight()+degree_sign+"</font>")
                     self.eightAMTempFortoday.setText("<font color='White'>"+Wapi.todayEightAM()+degree_sign+"</font>")   
                     self.fourPMTempFortoday.setText("<font color='White'>"+Wapi.todayFourtPM()+degree_sign+"</font>")   
                
                elif x=="Tomorrow":
                     self.midnighttemptextTime.setGeometry(40,20,50,20)  
                     self.midnighttemptextTime.setText("<font color='White'>00:00</font>")
                     self.eightAMtemptextTime.setGeometry(40,20,50,20)   
                     self.eightAMtemptextTime.setText("<font color='White'>08:00</font>")
                     self.fourPMtemptextTime.setGeometry(40,20,50,20)  
                     self.fourPMtemptextTime.setText("<font color='White'>16:00</font>")
                     self.midNightTempFortoday.setText("<font color='White'>"+Wapi.tomorrowMidNight()+degree_sign+"</font>")
                     self.eightAMTempFortoday.setText("<font color='White'>"+Wapi.tomorrowEightAM()+degree_sign+"</font>")   
                     self.fourPMTempFortoday.setText("<font color='White'>"+Wapi.tomorrowFourPm()+degree_sign+"</font>")  
        
                elif x=="Other":
                     
                     self.midnighttemptextTime.setGeometry(30,20,70,20)   
                     self.midnighttemptextTime.setText("<font color='White' size=0.1'>"+str((datetime.date.today() + datetime.timedelta(days=3)).strftime('%d-%m-%Y'))+"</font>")   
                     self.midNightTempFortoday.setText("<font color='White'>"+Wapi.dayAfterTomorrowWeather()+degree_sign+"</font>")
                     
                     self.eightAMtemptextTime.setGeometry(30,20,70,20)  
                     self.eightAMtemptextTime.setText("<font color='White' size=0.1'>"+str((datetime.date.today() + datetime.timedelta(days=4)).strftime('%d-%m-%Y'))+"</font>")    
                     self.eightAMTempFortoday.setText("<font color='White'>"+Wapi.fourthDayTomorrowWeather()+degree_sign+"</font>")   
                     
                     self.fourPMtemptextTime.setGeometry(30,20,70,20)  
                     self.fourPMtemptextTime.setText("<font color='White' size=0.1'>"+str((datetime.date.today() + datetime.timedelta(days=5)).strftime('%d-%m-%Y'))+"</font>") 
                     self.fourPMTempFortoday.setText("<font color='White'>"+Wapi.fifthDayTomorrowWeather()+degree_sign+"</font>") 
                        
                
               
#In this class we are calling OpenWeather API to get real time weather by giving that location            
class SingledayAPI():
        
        def __init__(self,city):
            self.city=city
            
            
        def urlAPI(self):
            url=('http://api.openweathermap.org/data/2.5/weather?q={}&appid=417ecd9eea84ffc6f0b4093c255446e2&units=metric'.format(self.city))
            res= requests.get(url)
            data=res.json()
            return data
            
        def tempNow(self):
            data=self.urlAPI()
            temp = str(data['main']['temp'])
            return temp   
        
        def windSpeed(self):
            data=self.urlAPI()
            return (str(data['wind']['speed']))
        
        def weatherType(self):
            data=self.urlAPI()
            print ((data['weather']['description']))
        
        def feel(self):
            data=self.urlAPI()
            
            return (str(data['main']['feels_like']))
        
        def humid(self):
            data=self.urlAPI()
            return (str(data['main']['humidity']))
        
        def get_pressure(self):
            data=self.urlAPI()
            return (str(data['main']['pressure']))
        

        def fiveDayAPI(self):
            url=('http://api.openweathermap.org/data/2.5/forecast?q={}&appid=417ecd9eea84ffc6f0b4093c255446e2&units=metric'.format(self.city))
            res= requests.get(url)
            data=res.json()
            return data
        
        def todaymidnight(self):
            data=self.fiveDayAPI()
            return (str(data['list'][0]['main']['temp']))
        
        def todayEightAM(self):
            data=self.fiveDayAPI()
            return (str(data['list'][2]['main']['temp']))
       
        def todayFourtPM(self):
            data=self.fiveDayAPI()
            return (str(data['list'][4]['main']['temp']))
        
        def tomorrowMidNight(self):
            data=self.fiveDayAPI()
            return(str(data['list'][6]['main']['temp']))
        
        def tomorrowEightAM(self):
            data=self.fiveDayAPI()
            return(str(data['list'][8]['main']['temp']))
        
        def tomorrowFourPm(self):
            data=self.fiveDayAPI()
            return(str(data['list'][10]['main']['temp']))
        
        def dayAfterTomorrowWeather(self):
            data=self.fiveDayAPI()
            return(str(data['list'][19]['main']['temp']))
        
        def fourthDayTomorrowWeather(self):
            data=self.fiveDayAPI()
            return(str(data['list'][26]['main']['temp']))
        
        def fifthDayTomorrowWeather(self):
            data=self.fiveDayAPI()
            return(str(data['list'][38]['main']['temp']))
            
        
def window():
    
    app = QApplication(sys.argv)
    win = myWindow()
    win.show()
    sys.exit(app.exec_())
        
        
window()        
```
